```python
def display_batch(batch):                                                                                                                                                                                   
    n, w, h = batch.shape                                                                                                                                                                                   
    final = batch.ravel().reshape((w * n, h)).T                                                                                                                                                             
    return Image.fromarray(np.vstack(final).astype('uint8'))   
```
